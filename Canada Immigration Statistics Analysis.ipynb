```python
df.groupby('Year')['Asylum-seekers'].sum()
```

```python
df.groupby('Year')['Asylum-seekers'].sum().plot(kind = 'bar')
plt.title('Total count of Asylum seekers in Canada by year')
plt.ylabel('Count of Asylum seekers')
```

```python
df.groupby('Year')['UNHCR-refugees'].sum()
```

```python
df.groupby('Year')['UNHCR-refugees'].sum().plot(kind = 'bar')
plt.title('Total count of Refugees in Canada by year')
plt.ylabel('Count of Refugees')
```

```python
df.groupby('Year')['total-count'].sum()
```

```python
df.groupby('Year')['total-count'].sum().plot(kind = 'bar')
plt.title('Total count of Refugees and Asylum seekers in Canada by year')
plt.ylabel('Count of Refugees and Asylum seekers')
```
